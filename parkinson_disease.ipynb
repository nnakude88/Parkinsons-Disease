## VISUALIZATION.

```python
plt.figure(figsize=(8, 6))
sns.histplot(pd['Age'], kde=True, color='blue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=pd, x='Gender', palette='coolwarm')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=pd, x='Ethnicity', palette='Set2')
plt.title('Ethnicity Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=pd, x='EducationLevel', palette='Set1')
plt.title('Education Level Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(pd['BMI'], kde=True, color='purple')
plt.title('BMI Distribution')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=pd, x='Smoking', palette='husl')
plt.title('Smoking Habits')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(pd['SystolicBP'], kde=True, color='green')
plt.title('Systolic Blood Pressure Distribution')
plt.xlabel('SystolicBP')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=pd, x='Hypertension', palette='dark')
plt.title('Hypertension Prevalence')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=pd, x='Gender', hue='Diagnosis', palette='Set2')
plt.title('Gender vs. Diagnosis')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Diagnosis')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=pd, x='EducationLevel', y='MoCA', palette='husl')
plt.title('Education Level vs. MoCA Scores')
plt.xlabel('Education Level')
plt.ylabel('MoCA Score')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=pd, x='Hypertension', y='SystolicBP', palette='Set3')
plt.title('Hypertension vs. Systolic Blood Pressure')
plt.xlabel('Hypertension')
plt.ylabel('Systolic BP')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=pd, x='Tremor', y='MoCA', palette='Paired')
plt.title('Tremor Severity vs. MoCA Scores')
plt.xlabel('Tremor')
plt.ylabel('MoCA Score')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=pd, x='FamilyHistoryParkinsons', hue='Diagnosis', palette='Set2')
plt.title('Family History of Parkinsons vs. Diagnosis')
plt.xlabel('Family History of Parkinsons')
plt.ylabel('Count')
plt.legend(title='Diagnosis')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=pd, x='Ethnicity', hue='Smoking', palette='Set2')
plt.title('Smoking by Ethnicity')
plt.xlabel('Ethnicity')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Smoking')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=pd, x='Smoking', hue='Gender', palette='Set1')
plt.title('Smoking Habits by Gender')
plt.xlabel('Smoking Status')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=pd, x='EducationLevel', hue='Smoking', palette='Set2')
plt.title('Smoking Habits by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Smoking')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=pd, x='Smoking', y='Age', hue='SleepDisorders', palette='Set2')
plt.title('Age Distribution by Smoking Status and Sleep Disorders')
plt.xlabel('Smoking Status')
plt.ylabel('Age')
plt.legend(title='Sleep Disorders')
plt.show()
```
